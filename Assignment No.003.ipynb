**3. Write a program that takes your full name as input and displays the abbreviations of the first and middle names except the last name which is displayed as it is. For example, if your name is Sunny Bhaveen Chandra, then the output should be S.B.Chandra.**

````python
text = input("Enter your name: ")
print(f"Enter your name : {text}")
a = text.split()
print(a[0][0]+'.'+ a[1][0]+'.' + a[2])
````

````python
text = input("Enter your name: ")
print(f"Enter your name : {text}")
first_name_start = text[0:1]+'.'
middle_name_start = text[7:8]+'.'
last_name = text[17:]
print(f"{first_name_start}{middle_name_start}{last_name}")
````

**6. Indian PAN card format follows the following formats** - 

    - `AYEPC7894X`
    - `ABCDE9999Y`
  Take user input for PAN_CARD and validate as per the above example.

````python
pan = input("Enter your PAN_No: ")
if pan[0:5].isalpha() and pan[0:5].isupper() and pan[5:9].isdigit() and pan[9].isalpha() and pan[9].isupper():
  print(f"User input for PAN_CARD :{pan} is valid.")
else:
  print(f"User input for PAN_CARD :{pan} is not valid.")
````
