##**2. Write a function which  will be able to print an index of list element without using an index function.**

```python
l = [1, 2, 3, 4, 5]
def indexEle(l):
  for i in range(len(l)):
    print(i, "====", l[i])

indexEle(l)
```

```python
def indexEle1(l):
  for i in range(0, findLength(l)):
    print(i, "===", l[i])

indexEle1(l)
```

```python
def indexEle2(l):
  l1 = []
  l2 = []
  for i in range(0, findLength(l)):
    l1.append(i)
    l2.append(l[i])
  return l1, l2

indexEle2(l)
```

##**5.Write a function which will take input as a list with any kind of numeric value and give an out as a multiplication of all the numeric data l = [3.5, 6.56, 4,5,"sudh" , "ineuron" , 'fsda bootcamp 2.0']**


```python
def multiplyList(l):

  result = 1
  for i in l:
    if type(i) == int or type(i) == float:
      result = result * i
  return result

l = [3.5, 6.56, 4, 5,  "sudh", "ineuron", "fsda bootcamp 2.0"]
multiplyList(l)
```
